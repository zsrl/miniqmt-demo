```python
from xtquant import xtdata
from datetime import datetime
import time

xtdata.connect(port=58609)

code_list = xtdata.get_stock_list_in_sector('沪深转债')

print('个数', len(code_list))

def on_data (datas):
    for key in datas.keys():
      readable_time = datas[key]['timetag']
      # timestamp_seconds = tick_time / 1000
      timestamp_seconds = datetime.strptime(readable_time, '%Y%m%d %H:%M:%S').timestamp()
      # 获取当前时间戳（秒级）
      current_timestamp_seconds = time.time()
      current_readable_time = datetime.fromtimestamp(current_timestamp_seconds).strftime('%Y-%m-%d %H:%M:%S')
      delay = current_timestamp_seconds - timestamp_seconds
      if (delay > 2):
          print(key, '数据时间', readable_time, '系统时间', current_readable_time, '延迟', delay) 
      

# seq = xtdata.subscribe_whole_quote(code_list=code_list, callback=on_data)

# xtdata.run()

datas = xtdata.get_full_tick(code_list)

print(datas)

on_data(datas)
```
